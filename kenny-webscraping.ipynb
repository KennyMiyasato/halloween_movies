# Master List Scrape
- Scrapes all articles above

```python
import requests
from bs4 import BeautifulSoup
import re
import csv

#######################
# Gets csv file ready #
#######################
csv_file = open('master_list.csv', 'w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['movie_name', 'year'])

#####################################
# Sets beautifulsoup site variables #
#####################################
page_imdb = requests.get("https://www.imdb.com/list/ls000091321")
soup2 = BeautifulSoup(page_imdb.content, 'html.parser')

page_townandcountrymag = requests.get("https://www.townandcountrymag.com/leisure/g12107335/best-classic-halloween-movies")
soup3 = BeautifulSoup(page_townandcountrymag.content, 'html.parser')

page_countryliving = requests.get("https://www.countryliving.com/life/entertainment/g3624/best-halloween-movies/?slide=35")
soup4 = BeautifulSoup(page_countryliving.content, 'html.parser')

page_timeout = requests.get("https://www.timeout.com/newyork/movies/best-halloween-movies-of-all-time")
soup5 = BeautifulSoup(page_timeout.content, 'html.parser')

page_esquire = requests.get("https://www.esquire.com/entertainment/movies/a38914/scary-movies-for-halloween")
soup6 = BeautifulSoup(page_esquire.content, 'html.parser')

###############
# imdb scrape #
###############
imdb_list_of_movies = soup2.find_all('div', class_='lister-item mode-detail')

for imdb_list_of_movie in imdb_list_of_movies:    
    movie_title = imdb_list_of_movie.find('div', class_='lister-item-content').h3.a.text
    print(f'Movie Name: {movie_title}')

    movie_year = imdb_list_of_movie.find('span', class_='lister-item-year').text
    print(f'Year: {movie_year}')
        
    print()
    
    # Filtering and cleaning our info
    if "(" in movie_year:
        movie_year = movie_year.replace("(", "").replace(")", "")
        
    if "I" in movie_year:
        movie_year = movie_year.strip("I").strip()
        
    if "TV" or "Video" or "â€“" in movie_year:
        continue

    
    csv_writer.writerow([movie_title, movie_year])

############################
# townandcountrymag scrape #
############################
townandcountrymag_list_of_movies = soup3.find_all('div', class_='listicle-slide')

for townandcountrymag_list_of_movie in townandcountrymag_list_of_movies:
    movie_title_and_year = townandcountrymag_list_of_movie.find('span', class_='listicle-slide-hed-text').text
    
    regexd_title = re.split("\((.+)\)", movie_title_and_year)[0]
    print(f'Movie Name: {regexd_title}')

    regexd_year = re.split("\W", movie_title_and_year)
    filtered_list = list(filter(None, regexd_year))
    movie_year = filtered_list[-1]
    print(f'Movie Year: {movie_year}')
    
    print()
    
    csv_writer.writerow([regexd_title, movie_year])

########################
# countryliving scrape #
########################
countryliving_list_of_movies = soup4.find_all('div', class_='slideshow-slide')

for countryliving_list_of_movie in countryliving_list_of_movies:
    
    try:
        movie_title_and_year = countryliving_list_of_movie.find('div', class_='slideshow-slide-hed').text
        
        regexd_title = re.split("\((.+)\)", movie_title_and_year)[0]
        
        regexd_year = re.split("\W", movie_title_and_year)
        filtered_list = list(filter(None, regexd_year))
        movie_year = filtered_list[-1]

    except Exception as e:
        regexd_title = None
        
    print(f'Movie Name: {regexd_title}')
    print(f'Movie Year: {movie_year}')
    print()
    
    csv_writer.writerow([regexd_title, movie_year])

##################
# Timeout scrape #
##################
timeout_list_of_movies = soup5.find_all('article', class_='listCard')

for timeout_list_of_movie in timeout_list_of_movies:
    
    movie_title_and_year = timeout_list_of_movie.find('div', class_='card-content').h3.a.text.strip()

    regexd_title = re.split("\((.+)\)", movie_title_and_year)[0]
    regexd_title.strip()

    regexd_year = re.split("\W", movie_title_and_year)
    filtered_list = list(filter(None, regexd_year))
    movie_year = filtered_list[-1]
    
    if "time" in movie_year:
        continue
    
    print(f'Movie Name: {regexd_title}')
    print(f'Movie Year: {movie_year}')
    print()

    csv_writer.writerow([regexd_title, movie_year])

##################
# Esquire scrape #
##################
esquire_list_of_movies = soup6.find_all('h4', class_='body-h4')
for esquire_list_of_movie in esquire_list_of_movies:
        
    try:
        movie_title = esquire_list_of_movie.find('em').text
        
    except Exception as e:
        pass
    
    print(f'Movie Name: {movie_title}')

    csv_writer.writerow([movie_title])

csv_file.close()
```

# IMDB
https://www.imdb.com/list/ls000091321

```python
import requests
from bs4 import BeautifulSoup
import re
import csv

page = requests.get("https://www.imdb.com/list/ls000091321")
soup2 = BeautifulSoup(page.content, 'html.parser')

csv_file = open('imdb_scrape.csv', 'w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['primary_key', 'movie_name', 'year', 'movie_mpaa_rating', 'regexd_year'])

imdb_list_of_movies = soup2.find_all('div', class_='lister-item mode-detail')

for imdb_list_of_movie in imdb_list_of_movies:
    imdb_primary_key_selector = imdb_list_of_movie.find('h3', class_='lister-item-header').a['href']
    imdb_primary_key = imdb_primary_key_selector.split('/')[2]
    print(f'PrimaryKey: {imdb_primary_key}')
    
    movie_title = imdb_list_of_movie.find('div', class_='lister-item-content').h3.a.text
    print(f'Movie Name: {movie_title}')

    movie_year = imdb_list_of_movie.find('span', class_='lister-item-year').text
    print(f'Year: {movie_year}')
    
    regexd_year = re.split("\W", movie_year)[1]
    print(f"Regex'd Year: {regexd_year}")
    
    # Try/Except when no mpaa rating exists
    try:
        movie_mpaa_rating = imdb_list_of_movie.find('span', class_='certificate').text

    except Exception as e:
        movie_mpaa_rating = None
    print(f'Rating: {movie_mpaa_rating}')
    
    print()
    
    csv_writer.writerow([imdb_primary_key, movie_title, movie_year, movie_mpaa_rating, regexd_year])

csv_file.close()
```
